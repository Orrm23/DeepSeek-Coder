```python
# prompt: create visual dashboard for speed,power,mean velocity,horsepower persecond analysis using rpm for atheletes speed analysis device where we calculate athelets speed through revolution per minute through absolute encoder sensor live.the dashboard should be more visual and should use 2d modeling for graphs of all above parameter and clearity in all above  graphs ploted.also build an python application for the same .write backend effective and correctly matching front end parameter

import matplotlib.pyplot as plt
import numpy as np
import time
import random
import pandas as pd
from IPython.display import clear_output

class AthleteAnalyzer:
  def __init__(self):
    self.rpm_data = []
    self.time_data = []
    self.power_data = []
    self.speed_data = []
    self.mean_velocity_data = []
    self.horsepower_per_second_data = []
    self.start_time = time.time()

  def calculate_speed(self, rpm, radius=0.3): # Radius in meters (adjust as needed)
    # Calculate linear speed (m/s) from RPM and radius
    circumference = 2 * np.pi * radius
    speed_mps = (rpm / 60) * circumference
    return speed_mps

  def calculate_power(self, rpm, torque=10): # Torque in Nm (adjust as needed)
    # Calculate power (watts) from RPM and torque
    power_watts = (rpm * 2 * np.pi * torque) / 60
    return power_watts

  def calculate_mean_velocity(self,speed_data):
    if len(speed_data)==0:
      return 0
    mean_velocity = sum(speed_data) / len(speed_data)
    return mean_velocity

  def calculate_horsepower_per_second(self,power_data):
    if len(power_data)==0:
      return 0

    horsepower_per_second = (power_data[-1] * 0.001341)
    return horsepower_per_second

  def update_data(self, rpm):
    current_time = time.time() - self.start_time
    self.time_data.append(current_time)
    self.rpm_data.append(rpm)

    speed = self.calculate_speed(rpm)
    self.speed_data.append(speed)

    power = self.calculate_power(rpm)
    self.power_data.append(power)

    mean_velocity=self.calculate_mean_velocity(self.speed_data)
    self.mean_velocity_data.append(mean_velocity)

    horsepower_per_second=self.calculate_horsepower_per_second(self.power_data)
    self.horsepower_per_second_data.append(horsepower_per_second)

  def create_dashboard(self):

    # Create figure and axes
    fig, axs = plt.subplots(3, 2, figsize=(18, 12))

    # RPM Graph
    axs[0, 0].plot(self.time_data, self.rpm_data, label="RPM", color="blue")
    axs[0, 0].set_xlabel("Time (s)")
    axs[0, 0].set_ylabel("RPM")
    axs[0, 0].set_title("RPM Over Time")
    axs[0, 0].grid(True)

    # Speed Graph
    axs[0, 1].plot(self.time_data, self.speed_data, label="Speed (m/s)", color="green")
    axs[0, 1].set_xlabel("Time (s)")
    axs[0, 1].set_ylabel("Speed (m/s)")
    axs[0, 1].set_title("Speed Over Time")
    axs[0, 1].grid(True)

    # Power Graph
    axs[1, 0].plot(self.time_data, self.power_data, label="Power (watts)", color="red")
    axs[1, 0].set_xlabel("Time (s)")
    axs[1, 0].set_ylabel("Power (watts)")
    axs[1, 0].set_title("Power Over Time")
    axs[1, 0].grid(True)

    #Mean Velocity Graph
    axs[1, 1].plot(self.time_data, self.mean_velocity_data, label="Mean Velocity", color="orange")
    axs[1, 1].set_xlabel("Time (s)")
    axs[1, 1].set_ylabel("Mean Velocity")
    axs[1, 1].set_title("Mean Velocity Over Time")
    axs[1, 1].grid(True)

    #Horse power per second Graph
    axs[2, 0].plot(self.time_data, self.horsepower_per_second_data, label="Horse Power Per Second", color="purple")
    axs[2, 0].set_xlabel("Time (s)")
    axs[2, 0].set_ylabel("Horsepower/s")
    axs[2, 0].set_title("Horse Power per Second Over Time")
    axs[2, 0].grid(True)

    fig.delaxes(axs[2,1])

    plt.tight_layout()
    plt.show()

  def run_analysis(self, duration=30, update_interval=1):
    end_time = time.time() + duration
    while time.time() < end_time:
      # Simulate RPM data from sensor (replace with actual sensor reading)
      rpm = random.randint(50, 150)
      self.update_data(rpm)
      self.create_dashboard()

      # Clear the output and update the dashboard
      clear_output(wait=True)

      time.sleep(update_interval)
    print("Analysis complete.")

if __name__ == "__main__":
  analyzer = AthleteAnalyzer()
  analyzer.run_analysis()

```
